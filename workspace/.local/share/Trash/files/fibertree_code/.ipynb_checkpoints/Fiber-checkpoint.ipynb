```python
import numpy as np
from fibertree import Tensor
# Convolution dimensions (Layer 15 example)
N, C, P, Q = 1, 1, 5, 5  # input/output shape
M, R, S    =  1, 3, 3       # filter shape
stride, pad = 1, 1
# Generate distinct input data
input_np = np.arange(N*C*P*Q).reshape(N, C, P, Q)
weight_np = np.arange(M*C*R*S).reshape(M, C, R, S)
output_np = np.zeros((N, M, P, Q), dtype=int)

# Simple convolution: output[n,f,p,q] += sum over c,r,s of I[n,c,h,w] * W[f,c,r,s]
for n in range(N):
    for m in range(M):
        for p in range(P):
            for q in range(Q):
                for c in range(C):
                    for r in range(R):
                        for s in range(S):
                            h = p * stride + r - pad
                            w = q * stride + s - pad
                            if 0 <= h < P and 0 <= w < Q:
                                output_np[n, c, p, q] += input_np[n, c, h, w] * weight_np[m, c, r, s]
# Convert to Fibertree Tensors
input_data = input_np.tolist()
weight_data = weight_np.tolist()
output_data = output_np.tolist()

# I = Tensor.fromUncompressed(
#     name="I",
#     rank_ids=["N","C","H","W"],
#     root=input_data,
#     shape=(N, C, P, Q),
#     color="blue"
# )
# W = Tensor.fromUncompressed(
#     name="W",
#     rank_ids=["M","C","R","S"],
#     root=weight_data,
#     shape=(M, C, R, S),
#     color="green"
# )

tm = TensorMaker()

N, C, P, Q = 1, 1, 5, 5  # input/output shape
M, R, S    =  1, 3,3       # filter shape

tm.addTensor("In", rank_ids=["N","C","P","Q"], shape=[N,C,P,Q], density=1, seed=40, color="blue")
tm.addTensor("W", rank_ids=["M","C","R","S"], shape=[M,C,R,S], density=1.0, seed=50, color="green")
tm.addTensor("O", rank_ids=["N","M","P","Q"], shape=[N,M,P,Q], density=0.0, seed=50, color="red")

# tm.displayControls()

A = tm.makeTensor("In")
B = tm.makeTensor("W")
C = tm.makeTensor("O")

N,C,P,Q = A.getShape()
M,C,R,S = B.getShape()
N,M,P,Q= C.getShape()
print(f"{M = }, {K = }")

displayTensor(A)
displayTensor(B)
displayTensor(C)



# O = Tensor.fromUncompressed(
#     name="O",
#     rank_ids=["N","M","P","Q"],
#     root=output_data,
#     shape=(N, M, P, Q),
#     color="red",
# )


# Display selected slice (e.g., first channel/filter)
# displayTensor(I)
# print()
# displayTensor(W)
# print()
# displayTensor(O)

# output_np = np.zeros((N, M, P, Q), dtype=int)
# output_data = output_np.tolist()
# O = Tensor.fromUncompressed(
#     name="O",
#     rank_ids=["N","M","P","Q"],
#     root=output_data,
#     shape=(N, M, P, Q),
#     color="red"
# )

```

```python
import numpy as np
from fibertree import Tensor
# Convolution dimensions (Layer 15 example)
N, C, P, Q = 1, 1, 5, 5  # input/output shape
M, R, S    =  1, 3,3       # filter shape
stride, pad = 1, 1
# Generate distinct input data
input_np = np.arange(N*C*P*Q).reshape(N, C, P, Q)
weight_np = np.arange(M*C*R*S).reshape(M, C, R, S)
output_np = np.zeros((N, M, P, Q), dtype=int)

# Simple convolution: output[n,f,p,q] += sum over c,r,s of I[n,c,h,w] * W[f,c,r,s]
for n in range(N):     
    for m in range(M):               
        for c in range(C):         
            for r in range(R):       
                for s in range(S):   
                    w_val = weight_np[m, c, r, s]          
                    for p in range(P):
                        for q in range(Q):
                            h = p*stride + r - pad
                            w = q*stride + s - pad
                            if 0 <= h < P and 0 <= w < Q:
                                output_np[n, m, p, q] += input_np[n, c, h, w] * w_val
# Convert to Fibertree Tensors
input_data = input_np.tolist()
weight_data = weight_np.tolist()
output_data = output_np.tolist()

I = Tensor.fromUncompressed(
    name="I",
    rank_ids=["N","C","H","W"],
    root=input_data,
    shape=(N, C, P, Q),
    color="blue"
)
W = Tensor.fromUncompressed(
    name="W",
    rank_ids=["M","C","R","S"],
    root=weight_data,
    shape=(M, C, R, S),
    color="green"
)
O = Tensor.fromUncompressed(
    name="O",
    rank_ids=["N","M","P","Q"],
    root=output_data,
    shape=(N, M, P, Q),
    color="red"
)

# Display selected slice (e.g., first channel/filter)
displayTensor(I)
print()
displayTensor(W)
print()
displayTensor(O)

```

```python
out_ref = C.getRoot() 
# print(C)
canvas  = createCanvas(A, B, C)
# I_k=I.getRoot()
# W_k=W.getRoot()
# canvas = createCanvas(,W,O)

for f, weight_f in B.getRoot():             # output‐filter
    for c, weight_c in weight_f:             # input‐channel
        for r, weight_r in weight_c:         # filter‐height
            for s, b_val in weight_r:        # filter‐width
                # b_val is now ‘stationary’
                for n, input_n in A.getRoot():       # batch
                    for c_in, input_c in input_n:    # channel
                        if c_in != c:                # only the matching channel
                            continue
                        for h, input_h in input_c:  # height
                            for w, a_val in input_h:# width
                                p = h - R + 1
                                q = w - S + 1
                                if 0 <= p < P and 0 <= q < Q:
                                    out_ref[n][f][p][q] += a_val * b_val
                                    canvas.addActivity(
                                        (n, c, h, w),    # input coords
                                        (f, c, r, s),    # weight coords
                                        (n, f, p, q)     # output coords
                                    )
displayCanvas(canvas)                                                          
```
